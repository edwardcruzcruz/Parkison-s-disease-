```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
theDataset = pd.read_csv('../neural_nets_experiments/results.csv')
theDataset.head()
```

```python
Dice = theDataset[theDataset['metric'] == 'Dice']
HD = theDataset[theDataset['metric'] == 'HD']
AVD = theDataset[theDataset['metric'] == 'AVD']
Recall = theDataset[theDataset['metric'] == 'Lesion detection']
F1 = theDataset[theDataset['metric'] == 'Lesion F1']
```

```python
ax1 = sb.barplot(x="patient", y="value", hue="Modelo aplicado", data=Dice).set_title('Dice (m√°s alto es mejor, max=1)')
plt.xlabel("N√∫mero de paciente")
plt.ylabel("Valor")
ax1
```

```python
ax2 = sb.barplot(x="patient", y="value", hue="Modelo aplicado", data=HD).set_title('HD (mm) (m√°s bajo es mejor, min=0)')
plt.xlabel("N√∫mero de paciente")
plt.ylabel("Valor")
ax2
```

```python
ax3 = sb.barplot(x="patient", y="value", hue="Modelo aplicado", data=AVD).set_title('AVD (%) (m√°s bajo es mejor, min=0)')
plt.xlabel("N√∫mero de paciente")
plt.ylabel("Valor")
ax3
```

```python
ax4 = sb.barplot(x="patient", y="value", hue="Modelo aplicado", data=Recall).set_title('Recall (m√°s alto es mejor, max=1)')
plt.xlabel("N√∫mero de paciente")
plt.ylabel("Valor")
ax4
```

```python
ax5 = sb.barplot(x="patient", y="value", hue="Modelo aplicado", data=F1).set_title('F1 (m√°s alto es mejor, max=1)')
plt.xlabel("N√∫mero de paciente")
plt.ylabel("Valor")
ax5
```

```python
fig1 = ax1.get_figure()
fig1.savefig('Dice_2.png')
fig2 = ax2.get_figure()
fig2.savefig('HD_2.png')
fig3 = ax3.get_figure()
fig3.savefig('AVD_2.png')
fig4 = ax4.get_figure()
fig4.savefig('Recall_2.png')
fig5 = ax5.get_figure()
fig5.savefig('F1_2.png')
```

```python
# Higher is better
#Improve (%) is the important
Dice2 = Dice.groupby('Modelo aplicado', as_index=False).mean()
value = Dice2[Dice2['Modelo aplicado'] == 'Estado del arte']['value']
Dice2['Improve (%)'] = Dice2.apply(lambda x: ((x['value'] - value) / value) * 100, axis=1) 
Dice2
```

```python
# lower is better
# Improvement (X times)
HD2 = HD.groupby('Modelo aplicado', as_index=False).mean()
value = HD2[HD2['Modelo aplicado'] == 'Estado del arte']['value']
HD2['Improvement (x times)'] = HD2.apply(lambda x: ((value  / x['value'])), axis=1)
HD2
```

```python
# lower is better
# value is the important
AVD2 = AVD.groupby('Modelo aplicado', as_index=False).mean()
value = AVD2[AVD2['Modelo aplicado'] == 'Estado del arte']['value']
AVD2['Performance'] = AVD2.apply(lambda x: ((value  - x['value'])), axis=1)
AVD2
```

```python
# higher is better
# Improve(%) is the important
Recall2 = Recall.groupby('Modelo aplicado', as_index=False).mean()
value = Recall2[Recall2['Modelo aplicado'] == 'Estado del arte']['value']
Recall2['Improve (%)'] = Recall2.apply(lambda x: ((x['value'] - value) / value) * 100, axis=1) 
Recall2
```

```python
# Higher is better
# Improve(%) is the important.
F1_2 = F1.groupby('Modelo aplicado', as_index=False).mean()
value = F1_2[F1_2['Modelo aplicado'] == 'Estado del arte']['value']
F1_2['Improve (%)'] = F1_2.apply(lambda x: ((x['value'] - value) / value) * 100, axis=1) 
F1_2
```
